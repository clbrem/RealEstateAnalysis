```python
from RealEstateAnalysis.data import DataCsv
from decimal import Decimal
from pprint import pprint as pp
import matplotlib.pyplot as plt
import numpy as np
import pandas
```

```python
import os
fileDir = os.path.dirname(os.path.realpath('__file__'))
fileName = os.path.join(fileDir,"RealEstateAnalysis/EBR_Building_Permits.csv")
print(fileName)
```

```python
def parseGeo(coords):
    parsed = coords.split("\n")[-1].replace("(", "").replace(")", "").split(",")
    if len(parsed) == 2:
        return [float(p) for p in parsed]
    else:
        return None
```

```python
def parseMoney(money):
    return float(money.replace("$", ""))
```

```python
data = DataCsv.new(fileName)
pp([b for a in data.load().sample(1).data for b in a.keys()])
```

```python
data.load() \
    .export(['subdivision']) \
    .dataFrame() \
    .drop_duplicates()
```

```python
df = data.process(geolocation=parseGeo) \
         .process(project_value=parseMoney) \
         .load() \
         .export(['zip','project_value']) \
         .dataFrame() 
df = df[df[1]>0]
df[1] = np.log(df[1])
df = df[np.isfinite(df[1])]
df.hist(by=0, bins=50, figsize=(20,15))
plt.show()

```

```python
it = data.process(geolocation=parseGeo) \
         .process(project_value=parseMoney) \
         .where(state="LA") \
         .load() \
         .export(['geolocation']) 
df = pandas.DataFrame([thing[0] for thing in it if thing[0] is not None])
df.plot(kind="scatter", x=1, y=0)



```
